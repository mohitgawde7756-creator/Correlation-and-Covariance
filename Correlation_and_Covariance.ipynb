2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?

-   Positive Covariance (Cov > 0)
Both variables move in the same direction.
Example: Income and expenditure.


Negative Covariance (Cov < 0)
Variables move in opposite directions.
Example: Product price and demand.


Zero Covariance (Cov = 0)
No linear relationship between the variables.
Changes in one variable do not systematically affect the other.

3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

-  Limitations of Covariance:

No fixed range, making interpretation difficult.
Depends on the units of measurement.
Cannot compare relationships across different datasets.
Indicates only direction, not strength of relationship.

Why Correlation is Preferred:

Has a fixed range from -1 to +1.
Unit-free measure.
Easy to interpret and compare.
Shows both strength and direction of the relationship.

Therefore, correlation is more useful and widely used than covariance.

4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?

-  | Basis        | Pearson's Correlation     | Spearman's Rank Correlation |
| ------------ | ------------------------- | --------------------------- |
| Data Type    | Numerical data            | Ranked or ordinal data      |
| Relationship | Linear relationship       | Monotonic relationship      |
| Assumptions  | Normally distributed data | Fewer assumptions           |
| Symbol       | r                         | ρ (rho)                     |


Use Spearman's Correlation when:

Data is in ranks.
Data is not normally distributed.
There are outliers.
Relationship is monotonic but not necessarily linear.

Example: Ranking students based on marks and attendance.

5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?

-  A correlation coefficient of 0.85 indicates:

A strong positive relationship between X and Y.
As X increases, Y tends to increase.
The variables move closely together.

Can causation be inferred?

No. Correlation does not imply causation because:

A third variable may influence both X and Y.
The relationship may be coincidental.
Correlation only measures association, not cause-and-effect.

Example: Ice cream sales and drowning incidents may have a high positive correlation because both increase during summer, but ice cream sales do not cause drowning.

Conclusion: A correlation of 0.85 shows a strong positive association, but it does not prove that one variable causes the other.
